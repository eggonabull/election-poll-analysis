## Data choice

I chose to use the most recent polls for each state available on 270togo.com

```python
import pandas as pd

x1 = pd.ExcelFile('presidential-polls.xlsx')
parsed = x1.parse()[1:101]

parsed
```

## Getting the data into an easier form to handle

I want both the clinton and trump predictions on the same row so that I'm not moving around in the document to get all the data together


```python
from electoral import states
parsed["join_index"] = parsed.index + 1
odd_rows = parsed[parsed.index % 2 == 1]
grouped = odd_rows.join(parsed, "join_index", rsuffix=" second")
def get_clinton(row):
    return row["Unnamed: 5"] if row["Results"] == "Clinton" else row["Unnamed: 5 second"]
def get_trump(row):
    return row["Unnamed: 5"] if row["Results"] == "Trump" else row["Unnamed: 5 second"]
def get_seats(row):
    return states[row["Location"]]
grouped["Clinton"] = grouped.apply(get_clinton, axis=1)
grouped["Trump"] = grouped.apply(get_trump, axis=1)
grouped["Seats"] = grouped.apply(get_seats, axis=1)

data = grouped.loc[:,("Date", "Poll Source", "Location", "Clinton", "Trump", "Seats")]
data

```

## Plan of attack

Okay so we've got our data, let's talk about how we can make some predictions off of this data that do not introduce too much bias not suggested by the data.


Looking at the polls for each state, if you imagine that the trump voters are stacked on the top and the cliton votes are stacked up from the bottom, with the undecideds in-between, then an argument can be made for a very simple model that positions the *actual vote* distributed around the average between the two.

For this project, I'm going to use a normal distribution because it's easy and I'm lazy.

For every state, I'm going to create a normal distribution from 0 to 100.  The distribution will be centered between the clinton vote and the trump vote, and the std dev will be half the number of undecided votes plus 3.

![Normal Distribution setup](./clinton-trump-shitty-normal-dist.png)



```python
# Add the centers
def get_center(row):
    return row["Clinton"] + (100 - row["Trump"] - row["Clinton"]) / 2
data["Center"] = data.apply(get_center, axis=1)

# add the deviations
def get_deviation(row):
    return (abs(row["Clinton"] - row["Center"]) / 2) + 3
data["Stddev"] = data.apply(get_deviation, axis=1)
data
```

## Self-analysis

This distributes the probability evenly around the polling data.  No predictive bullshit.  None o' that shit. Predictive models are bullshit.  That's what the conservative elite data scientists tell me. 

It also assumes that everybody votes and that all states allocate all of their electors to the majority popular vote. (sorry maine and nebraska, I love your voting systems but ain't nobody got time for that shit)

It also increases in uncertainty as the number of undecided voters grows.  That seems reasonable. 

So I'm not smart enough to generate a full picture of all the possible electoral outcomes with their probabilities. Whatever, let's just jam through 10,000 elections and see kind of what plays out.

```python
from numpy.random import normal


def run_election(times):
    result_frame = pd.DataFrame()

    def get_seats(row):
        result = normal(row["Center"], row["Stddev"])
        return row["Seats"] if result > 50 else 0

    for i in range(times):
        result_frame[i] = data.apply(get_seats, axis=1)

    return result_frame
    

# given an example of the output
run_election(20)
```

```python
election_results = run_election(10000)
hillary_wins = election_results.apply(lambda x: 1 if sum(x) > 270 else 0)
hillary_wins.sum()

# give this a while to run, the way this is calculated is 
# stupid inefficient, but whatever the computer is a
# harder worker than I am 
```
